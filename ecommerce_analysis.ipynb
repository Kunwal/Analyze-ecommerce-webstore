Over all Analyze

```python
repeated_customer = df['customer_id'].value_counts()
```

```python
overall_path = 'output_analyze/overall_analyze.txt'
```

```python
import os

# Create the folder if it doesn't exist
os.makedirs("output_analyze", exist_ok=True)
```

```python
# Total sales 
repeating_customers = (repeated_customer >= 2).sum()
with open(overall_path, 'w') as f:
    f.write(f' {'-'*50}\n Total repeating customers on my store is: {repeating_customers}\n {'-'*50}')
```

```python
repeated_product = df['product_name'].value_counts()
# repeated_product
```

```python
# top selling product
top_products = repeated_product.iloc[0:3]
with open(overall_path, 'a') as f:
     f.write(f'\n {'-'*30}\n Top 3 selling product\n {'-'*30}')
     for name, count in top_products.items():
        f.write(f"\n {name}: {count}")
```

```python
# Least selling product
least_product = repeated_product.iloc[-3:]
with open(overall_path, 'a') as f:
     f.write(f'\n {'-'*30}\n Least 3 selling product\n {'-'*30}')
     for name, count in least_product.items():
        f.write(f"\n {name}: {count}")
```

```python
# Repeated customers more than 5 times
with open(overall_path, 'a') as f:
    f.write(f'\n {'-'*50}\n Repeated customers more than 5 times\n {'-'*50}')
    for index ,value in repeated_customer.items():
            if value >= 5:
               name = df.loc[df['customer_id'] == index, 'name'].iloc[0]
               f.write(f"\n {name}: {value}") 
```

```python
# total income
totalprice = (df['quantity']*df.price).sum()
with open(overall_path, 'a') as f:
    f.write(f'\n {'-'*50}\n Total Income: {totalprice} \n {'-'*50}')
```

Yearly Analyze

```python
df['year'] = df.order_date.dt.year
```

```python
year_group = df.groupby('year')
```

```python
yealy_path = 'output_analyze/yearly_analyze.txt'
```

```python
# Yearly Total repeating customers
with open(yealy_path, 'w') as f:
    f.write(f' {'-'*50}')
    for year in df.year.unique():
        yearwisegroup = year_group.get_group(year)
        year_repeated_customer= yearwisegroup.customer_id.value_counts()
        year_repeating_customers = (year_repeated_customer >= 2).sum()
        f.write(f'\n Total repeating customers on my store is {year}: {year_repeating_customers} \n {'-'*50}')
```

```python
# Top selling product
for year in df.year.unique():
    yearwisegroup = year_group.get_group(year)
    year_repeated_product = yearwisegroup['product_name'].value_counts()
    year_top_products = year_repeated_product.iloc[0:3]
    with open(yealy_path, 'a') as f:
        f.write(f'\n {'-'*50}\n Top 3 selling products in {year} \n {'-'*50}')
        for name, count in year_top_products.items():
            f.write(f"\n {name}: {count}")
```

```python
# Least selling product
for year in df.year.unique():
    yearwisegroup = year_group.get_group(year)
    year_repeated_product = yearwisegroup['product_name'].value_counts()
    year_least_products = year_repeated_product.iloc[-3:]
    with open(yealy_path, 'a') as f:
        f.write(f'\n {'-'*50}\n Least selling products in {year} \n {'-'*50}')
        for name, count in year_least_products.items():
            f.write(f"\n {name}: {count}")
```

```python
# Repeated customers more than 5 times
for year in df.year.unique():
    yearwisegroup = year_group.get_group(year)
    year_repeated_customer= yearwisegroup.customer_id.value_counts()
    with open(yealy_path, 'a') as f:
        f.write(f'\n {'-'*50}\n Repeated customers more than 5 times {year} \n {'-'*50}')
        for index ,value in year_repeated_customer.items():
                if value >= 5:
                   name = yearwisegroup.loc[yearwisegroup['customer_id'] == index, 'name'].iloc[0]
                   f.write(f"\n {name}: {value}") 
```

```python
# Yearly total income
with open(yealy_path, 'a') as f:
    f.write(f'\n{'-'*50}')
    for year in df.year.unique():
        yearwisegroup = year_group.get_group(year)
        totalprice = (yearwisegroup['quantity']*yearwisegroup.price).sum()
        f.write(f'\nIn {year} Total sales: {totalprice} \n{'-'*50}')
    # print(f'In {year} Total sales: {totalprice}')
```

Monthly Analyze

```python
df['month'] = df.order_date.dt.month_name()
```

```python
monthly_path = 'output_analyze/monthly_analyze.txt'
```

```python
with open(monthly_path, 'w') as f:
    for year in df['year'].unique():
        yearwisegroup = year_group.get_group(year)
        f.write(f'\n{'-'*50}\n Total repeating customers in {year} \n{'-'*50}')
    
        month_group = yearwisegroup.groupby('month')
        for month in yearwisegroup['month'].unique():
            monthwisegroup = month_group.get_group(month)
            
            # Count customers who appear more than once
            customer_counts = monthwisegroup['customer_id'].value_counts()
            repeating_customers = (customer_counts >= 2).sum()
            
            f.write(f"\nTotal repeating customers in {month}: {repeating_customers}")
```

```python
# Monthly Top selling product
with open(monthly_path, 'a') as f:
    for year in df['year'].unique():
        yearwisegroup = year_group.get_group(year)
        f.write(f"\n{'='*40}\n Top selling product on my store {year} \n{'='*40}")
        month_group = yearwisegroup.groupby('month')
        for month in yearwisegroup['month'].unique():
            monthwisegroup = month_group.get_group(month)
            month_repeated_product = monthwisegroup['product_name'].value_counts()
            month_top_products = month_repeated_product.iloc[0:3]
            f.write(f'\n{'-'*40}\n Top 3 selling products in {month}\n{'-'*40}')
            for name, count in month_top_products.items():
                f.write(f"\n{name}: {count} ")
```

```python
# Monthly least selling product  
with open(monthly_path, 'a') as f:
    for year in df['year'].unique():
        yearwisegroup = year_group.get_group(year)
        f.write(f"\n{'='*40}\n least selling product on my store {year} \n{'='*40}")
        month_group = yearwisegroup.groupby('month')
        for month in yearwisegroup['month'].unique():
            monthwisegroup = month_group.get_group(month)
            month_repeated_product = monthwisegroup['product_name'].value_counts()
            month_least_products = month_repeated_product.iloc[-3:]
            f.write(f'\n{'-'*40}\n least 3 selling products in {month}\n{'-'*40}')
            for name, count in month_least_products.items():
                f.write(f"\n{name}: {count} ")
```

```python
# Monthly Repeated customers more than 5 times
with open(monthly_path, 'a') as f:
    for year in df['year'].unique():
        yearwisegroup = year_group.get_group(year)
        f.write(f"\n{'='*50}\n Repeated customers more than 5 times {year} \n{'='*50}")
        month_group = yearwisegroup.groupby('month')
        for month in yearwisegroup['month'].unique():
            monthwisegroup = month_group.get_group(month)
            month_repeated_customer= monthwisegroup.customer_id.value_counts()
            for index ,value in month_repeated_customer.items():
                if value >= 5:
                    name = monthwisegroup.loc[monthwisegroup['customer_id'] == index, 'name'].iloc[0]
                    f.write(f"\n{name}: {value} ")
```

```python
# Monthly total income
with open(monthly_path, 'a') as f:
    for year in df['year'].unique():
        yearwisegroup = year_group.get_group(year)
        f.write(f"\n{'-'*50}\n Monthly sales in {year} \n{'-'*50}")
        month_group = yearwisegroup.groupby('month')
        for month in yearwisegroup['month'].unique():
            monthwisegroup = month_group.get_group(month)
            totalprice = (monthwisegroup['quantity']*monthwisegroup.price).sum()
            f.write(f'\n In {month} sales: {totalprice}')
```
